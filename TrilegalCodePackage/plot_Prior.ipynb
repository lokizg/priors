```python
import numpy as np
import LocusTools as lt
import PlotTools as pt
import param_tool as param_tool
from astropy.io import ascii
```

```python
fitting_params = param_tool.getFittingParams()

rmagMin = fitting_params['rmag_MIN']
rmagMax = fitting_params['rmag_MAX']
rmagBinSZ = fitting_params['r_BinSize']
rmagNsteps = fitting_params['rmag_Nsteps']
```

```python
ra = 188
dec = 0.1
nside = 64 #Temporarily, only 64 is available

rootname = param_tool.getPrior(ra, dec, nside, rmagMin, rmagMax, rmagBinSZ, rmagNsteps)
```

```python
# Choose the map index for the magnitude range you want; ex: 17 for r=22.0-23.0
map_ind = 13
mid_rMag = rmagMin + rmagBinSZ*map_ind
prior = param_tool.getMap(map_ind, rootname)
```

```python
# plot one map
pt.restore2Dmap(prior, mid_rMag, logScale=True, save_plot=False)
```
