```python
import healpix_prob as healpix_prob
import numpy as np
```

```python
# Assign Healpix resolution, n_side, Healpixel, healpix_num, and r-band bin size, and creating a healsky as an object
heal_star = healpix_prob.Stars(n_side = 64, healpix_num = 35699, r_binsize = 1.0)
```

```python
# It reads data from TRILEGAL catalog and convert some values in Mr and Ar 
heal_star.read_data()
```

```python
# Give a value of r-band magnitude and select stars within the r-band bin where the given r-band mag is in
heal_star.select_star(18.3)
```

```python
# Get fitting result for Mr
mr_fitting = heal_star.mr_fitting()
```

```python
# Get Mr probability distribution function according to the given sample bins
mr_sample_bins = 0.25*np.arange(49) - 6
mr_pdf = heal_star.mr_prob(mr_fitting, mr_sample_bins)
```

```python
# One can check Mr probability distribution function 
heal_star.plot_mr(mr_pdf, mr_sample_bins, filepath = "/astro/users/danichao/mr_fitting_test.pdf")
```

```python
# Get fitting result for FeH
feh_fitting = heal_star.feh_fitting()
```

```python
# Get FeH probability distribution function according to the given sample bins
feh_sample_bins = 0.125*np.arange(31) - 3
feh_pdf = heal_star.feh_prob(feh_fitting, feh_sample_bins)
```

```python
# Visualize fitted FeH result according the given sample bins
heal_star.plot_feh(feh_pdf, feh_sample_bins, filepath = "/astro/users/danichao/feh_fitting_test.pdf")
```

```python
# Get fitting result for Ar
ar_fitting = heal_star.ar_fitting()
```

```python
# Get Ar probability distribution function according to the given sample bins
ar_sample_bins = 0.125*np.arange(41)
ar_pdf = heal_star.ar_prob(ar_fitting, ar_sample_bins)
```

```python
# Visualize fitted Ar result 
heal_star.plot_ar(ar_pdf, ar_sample_bins, filepath = "/astro/users/danichao/ar_fitting_test.pdf")
```
